# # Using torch

```
import torch
```

```
seq = torch.nn.Sequential(torch.nn.Linear(5,5), torch.nn.Dropout(0.5), torch.nn.Linear(5,1))
```

```
print(seq)
```

```
seq_linear = torch.nn.Sequential(torch.nn.Linear(5,5), torch.nn.Dropout(0), torch.nn.Linear(5,1))
```

```
print(seq_linear)
```

```
x_input = torch.rand(5,5)
```

```
x_input
```

```
x = x_input

for layer in seq:
     x = layer( x )
     print(x)
```

```
x = x_input

for layer_linear in seq_linear:
     x = layer_linear(x)
     print(x)
```
